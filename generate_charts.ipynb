```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("EDGA_dataset.csv", encoding='utf-8')

indexForCountry =  dict(zip( list(df["Country"]),list(df.index)))
```

```python
indexForCountry
```

```python


def getPayeChart(name):
    x = df.iloc[indexForCountry[name]].values[1:]
    plt.figure(figsize=(10, 6))
    plt.plot(x.T)

    # Adding labels and title
    plt.title('Scatter Plot with Trend Line - Temperature by Year')
    plt.xlabel('Year')
    plt.ylabel('Temperature (°C)')

    # Display the legend
    plt.legend()
    plt.grid(True)
    plt.savefig('temp.png')
    # Display the plot
    plt.show()  

getPayeChart("Tunisia")
```
